```python
#Задание 1
#Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
#- оценка 2 и меньше - низкий рейтинг
#- оценка 4 и меньше - средний рейтинг
#- оценка 4.5 и 5 - высокий рейтинг

#Результат классификации запишите в столбец class
```

```python
import pandas as pd
```

```python
data = pd.read_csv('/Users/max/Downloads/ml-20m/ratings.csv')
data.head()
```

```python
def classification(a):
    "Классификатор рейтинга"
    if a <= 2:
        return 'низкий рейтинг'
    elif a > 2 and a <=4:
        return 'средний рейтинг'
    elif a == 4.5 or a == 5:
        return 'высокий рейтинг'
```

```python
data['class'] = data['rating'].apply(classification)
data.tail(10)
```

```python
dt = pd.read_csv('keywords.csv')
dt.tail()
```

```python
geo_data = {

'Центр': ['москва','тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def geoclass(row):
    cities = []
    for city in list(geo_data.values()):
        cities +=city
    for city in cities:
        if city in row:
            for key in geo_data:
                if city in geo_data[key]:
                    return key
    else: return 'undefined'
```

```python
dt['region'] = dt['keyword'].apply(geoclass)
dt.head()
dt.region.unique()
```

```python
filter = dt[(dt['region'] == 'Дальний Восток')]
filter.tail()
```

```python
movies = pd.read_csv('/Users/max/Downloads/Python_7_pandas_functions_groupby/ml-latest-small/movies.csv')
movies.head()
```

```python
years = list(range(1950,2011))
```

```python
def production_year(line):
    #years = list(range(1950,2011))
    for year in years:
        if str(year) in line:
            return year
    else: return '1900'
```

```python
movies['year'] = movies['title'].apply(production_year)
movies.tail()
```

```python
ratings = pd.read_csv('/Users/max/Downloads/Python_7_pandas_functions_groupby/ml-latest-small/ratings.csv')
ratings.head()
```

```python
result = pd.merge(movies, ratings, on='movieId', how='outer')
result.head()
```

```python
def avarage_rating(result):
    return result.rating.mean()
```

```python
final = result.groupby('year').apply(avarage_rating)
```
