#Задание 1
Посчитайте среднее время жизни пользователей, 
которые выставили более 100 оценок. Под временем жизни
понимается разница между максимальным и минимальным значением столбца timestamp
для данного значения userId

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
raitings_grouped = ratings.groupby('userId').count().reset_index()
filtered_ratings = raitings_grouped[raitings_grouped['rating'] >=100]
film_fans_user_ids = filtered_ratings['userId'].tolist()
```

```python
for user in film_fans_user_ids:
    avarage_timestamp = max((ratings[ratings['userId'] == user]['timestamp'])) - min((ratings[ratings['userId'] == user]['timestamp']))
    print(f' userId: {user} avarage_timestamp: {avarage_timestamp}')
```

Задача 3 В задаче сквозной аналитики вам предоставили данные по местоположению пользователей. Т. е. для каждого user_id известна последовательность координат (широта/долгота), когда они требовались приложению для полноценной работы. Как бы вы добавили эти сведения в таблицу визитов и покупок?


Логично полагать, что главная связь координат пользователя должна быть с его визитами, т.е. пользователь зашел в приложение - фиксируются его координаты.



```python
visits = pd.DataFrame(
    {
        'user_id': [11, 22, 55, 11, 77],
        'source': ['ad', 'yandex', 'email', 'google', 'ad']
    }
)

visits = visits[['user_id', 'source']]
visits
```

```python
visits_grouped = visits.groupby('user_id').count().reset_index()
visits_grouped.rename(columns={'source': 'visits'}, inplace=True)
visits_grouped
```

```python
purchases = pd.DataFrame(
    {
        'user_id': [11, 22, 55, 11, 99],
        'category': ['Спорт', 'Авто', 'Дача', 'Спорт', 'Авто'],
    }
)

purchases = purchases[['user_id', 'category']]
purchases
```

```python
locations = pd.DataFrame(
    {
        'user_id': [11, 22, 55, 11, 99],
        'location': ['xyz', 'yzx', 'zxy','zzz', 'zzy'],
    }
)

locations = locations[['user_id', 'location']]
locations
```

```python
df = locations.merge(purchases, on='user_id', how='left')
df1 = df.drop_duplicates()
df1
```

```python
df1.pivot_table(index='location', columns='category', fill_value = 0).reset_index()
```
